```python
import matplotlib.pyplot as plt
import numpy as np

residues = ["K224", "K225", "K274", "K275"]
with_pip2 = [144, 140, 54, 159]
with_pops = [26, 11, 61, 85]

fig, ax = plt.subplots(figsize=(10, 6))

bar_width = 0.30
gap_width = 0.02  # Adjust this value for the desired gap
opacity = 0.9

x = np.arange(len(residues))

bar1 = ax.bar(x - bar_width/2 - gap_width/2, with_pip2, bar_width, alpha=opacity, color='purple', label="With PIP2")
bar2 = ax.bar(x + bar_width/2 + gap_width/2, with_pops, bar_width, alpha=opacity, color='brown', label="With POPS")

ax.set_xticks(x)
ax.set_xticklabels(residues, fontsize=20, rotation=0, ha='center')

# Add an additional x-axis label below the existing label
ax.set_xlabel('Residue', fontsize=20)
ax.set_xlabel('Loop1 Residues                Loop2 Residues', fontsize=22, labelpad=20)

ax.set_ylabel('H-bond Occupancy (%)', fontsize=22)
ax.tick_params(axis='both', labelsize=22)
ax.legend(fontsize=18, frameon=False, loc='upper center')

# Display every other y-axis tick label
for index, label in enumerate(ax.yaxis.get_ticklabels()):
    if index % 2 == 0:
        label.set_visible(True)
    else:
        label.set_visible(False)

# Add a little gap between the bars
plt.subplots_adjust(wspace=0.3)

# Add "%" symbol and "Occupancy" in the second line of the y-axis label
ax.set_ylabel('% Of H-bond \n Occupancy', fontsize=22)

# Add an imaginary vertical black line in the middle
#ax.axvline(x=len(residues) / 2.75, color='black', linestyle='--')
# Adjust the vertical line position to extend to the middle and a bit below the bottom
ax.axvline(x=len(residues) / 2.75, color='black', linestyle='--', ymin=-0, ymax=-0.1)
#ax.axvline(y=-20, color='black', linestyle='--')

plt.tight_layout()
plt.savefig('HPercentage.png', dpi=300, bbox_inches='tight')
plt.show()
```
